```
import numpy as np 
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score,r2_score,jaccard_similarity_score,classification_report
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
```

```
np.random.seed(0)
```

```
x,y = make_blobs(n_samples = 50000,centers =[[4,4],[-2,-1],[2,-3]],cluster_std=0.9)
y
```

```
plt.scatter(x[:,0],x[:,1],marker = '.')
```

```
k_mean= KMeans(init ="k-means++" ,n_clusters=3 ,n_init =12)
```

```
k_mean.fit(x)
```

```
y_hat = k_mean.predict(x)
```

```
y_hat
```

```
k_mean_center= k_mean.cluster_centers_
```

```
kmean_labels = k_mean.labels_
```

```
k_mean_center,kmean_labels
```

```
fig = plt.figure(figsize = (6,4))
```

```
colors = plt.cm.Spectral(np.linspace(0,1,len(kmean_labels)))
```

```
ax = fig.add_subplot(1,1,1)
```

```
for k,col in zip(range(len([[4,4],[-2,-1],[2,-3]])),colors):
  my_members = (kmean_labels == k )
  cluster_centre = k_mean_center[k]
  plt.plot(x[my_members,0],x[my_members,1],'w',markerfacecolor= col,marker = '.')
  plt.plot(cluster_centre[0],cluster_centre[1],'o',markerfacecolor = col,markeredgecolor = 'k',markersize = 6)

plt.show()
```
