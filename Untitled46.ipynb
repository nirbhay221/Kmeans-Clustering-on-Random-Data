```
import numpy as np
import pandas as pd 
import sklearn 
import matplotlib.pyplot as plt 
from sklearn import preprocessing
from sklearn.cluster import DBSCAN
from sklearn.datasets.samples_generator import make_blobs
from sklearn.model_selection import train_test_split
```

```
def CreateDataPoints(centroidLocation,numSamples,clusterDeviation):
  x,y= make_blobs(n_samples = numSamples,centers = centroidLocation,cluster_std=clusterDeviation)
  x = preprocessing.StandardScaler().fit(x).transform(x)
  return x,y
```

```
x, y = CreateDataPoints([[4,3], [2,-1], [-1,4]] , 1500, 0.5)
```

```
epsillon = 0.3
minimumsamples = 7
db = DBSCAN(eps = epsillon,min_samples= minimumsamples).fit(x)
labels = db.labels_
```

```
core_samples_mask = np.zeros_like(db.labels_,dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
core_samples_mask
```

```
n_clusters = len(set(labels)) -(1 if -1 in labels else 0)
n_clusters
```

```
unique_labels = set(labels)
unique_labels
```

```
colors = plt.cm.Spectral(np.linspace(0,1,len(unique_labels)))
for k, col in zip(unique_labels, colors):
  if k == -1:
    col='k'
  class_member = (labels ==k)
  xy = x[class_member & core_samples_mask]
  plt.scatter(xy[:,0],xy[:,1],s=50,c=[col],marker =u'o',alpha = 0.5)
  xy = x[class_member & ~core_samples_mask]
  plt.scatter(xy[:,0],xy[:,1],s=50,c=[col],marker =u'o',alpha = 0.5)
```

```
from sklearn.cluster import KMeans

kmeans = KMeans(init= "k-means++",n_clusters =3,n_init=12)
kmeans.fit(x)
labels = kmeans.labels_
k = 3
figure = plt.figure(figsize=(6,4))
ax = figure.add_subplot(1,1,1)
for k,col in zip(range(k),colors):
  class_mem = (kmeans.labels_ == k)
  plt.scatter(x[class_mem,0],x[class_mem,1],c =col,marker = u'o',alpha = 0.5) 
plt.show()
```
